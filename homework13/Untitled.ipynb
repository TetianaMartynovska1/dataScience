```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import AgglomerativeClustering
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, 
                           cluster_std = 0.6, random_state = 0)
colors = np.array([plt.cm.Spectral(val)
          for val in np.linspace(0, 1, len(set(y)))])
plt.figure(figsize=(8,6))

plt.scatter(X[:,0], X[:,1], c= colors[y], s= 20) 
```

```python
plt.show()
```

```python
from scipy.cluster.hierarchy import complete, average, linkage, single, dendrogram
```

```python
model = AgglomerativeClustering(n_clusters=5, linkage='complete')
labels = model.fit_predict(X)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set1', s=100, edgecolors='k')
plt.title(' Complete')
plt.xlabel('X')
plt.ylabel('Y')
plt.tight_layout()
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=5, linkage='single')
labels = model.fit_predict(X)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set2', s=100, edgecolors='k')
plt.title(' single')
plt.xlabel('X')
plt.ylabel('Y')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
clusters = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set2', s=100, edgecolors='k')
plt.title(' DBSCAN')
plt.xlabel('X')
plt.ylabel('Y')
plt.tight_layout()
plt.show()
```
