```python
gdp_years = ['Country'] + [str(year) for year in range(2006, 2016)]
GDP = GDP[gdp_years]
```

```python
GDP
```

```python
ScimEn_top15 = ScimEn[ScimEn['Rank'] <= 15]
```

```python
ScimEn_top15
```

```python
merged_data = pd.merge(GDP, ScimEn_top15[['Rank', 'Country', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index']], on='Country', how='inner')
```

```python
answer_one = pd.merge(merged_data, energy, on='Country', how='inner')
```

```python
answer_one
```

```python
def answer_two():
    Top15 = answer_one
    answer_one['Average GDP'] = answer_one[[str(year) for year in range(2006, 2016)]].mean(axis=1, skipna=True)
    return answer_one[['Country', 'Average GDP']].sort_values(by=['Average GDP']).iloc[::-1]
```

```python
answer_two()
```

```python
def answer_three():
    Top15 = answer_two()
    country_6th = Top15.iloc[5]
    country_name = country_6th['Country']
    
    gdp_2006 = answer_one.loc[answer_one['Country'] == country_name, '2006'].values[0]
    gdp_2015 = answer_one.loc[answer_one['Country'] == country_name, '2015'].values[0]

    gdp_change = gdp_2015 - gdp_2006
    
    return pd.DataFrame({'Country': [country_name], 'GDP Change': [gdp_change]})
```

```python
answer_three()
```

```python
def answer_four():
    Top15 = answer_one
    Top15["Self-Citation Ratio"] = Top15["Self-citations"] / Top15["Citations"]
    max_country = Top15.loc[Top15["Self-Citation Ratio"].idxmax()]
    return (max_country["Country"], max_country["Self-Citation Ratio"])
```

```python
answer_four()
```

```python
def answer_five():
    Top15 = answer_one
    Top15['Population'] = Top15['Energy Supply']/Top15['Energy Supply per Capita']
    return Top15[['Country', 'Population']].sort_values(
        by='Population', ascending=False).iloc[2]
```

```python
answer_five()
```

```python
def answer_six():
    Top15 = answer_one.copy()
    Top15["CD per capita"] = Top15["Citable documents"] / Top15["Population"]
    return Top15["CD per capita"].corr(Top15["Energy Supply per Capita"])
```

```python
answer_six()
```

```python
def answer_seven():
    Top15 = answer_one
    
    ContinentDict  = {
        'China': 'Asia', 
        'United States': 'North America', 
        'Japan': 'Asia', 
        'United Kingdom': 'Europe', 
        'Russian Federation': 'Europe', 
        'Canada': 'North America', 
        'Germany': 'Europe', 
        'India': 'Asia',
        'France': 'Europe', 
        'South Korea': 'Asia', 
        'Italy': 'Europe', 
        'Spain': 'Europe', 
        'Iran': 'Asia',
        'Australia': 'Australia', 
        'Brazil': 'South America'
    }
    
    Top15['Continent'] = Top15.index.map(ContinentDict)
    
    result = Top15.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])
    
    return result
    print(Top15[['Continent']].drop_duplicates())
```
